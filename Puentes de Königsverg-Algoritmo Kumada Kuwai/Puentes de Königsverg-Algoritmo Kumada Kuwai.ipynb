# Problema de los puentes de K√∂nigsberg



El problema de los puentes de K√∂nigsberg, tambi√©n llamado m√°s espec√≠ficamente problema de los siete puentes de K√∂nigsberg, es un c√©lebre problema matem√°tico resuelto por Leonhard Euler en 1736 y cuya resoluci√≥n dio origen a la teor√≠a de grafos. Su nombre se debe a K√∂nigsberg, la ciudad de Prusia Oriental y luego de Alemania que desde 1945 se convirti√≥ en la ciudad rusa de Kaliningrado.

Esta ciudad est√° atravesada por el r√≠o Pregolia. Este se bifurca y rodea con sus brazos a la isla Kneiphof,de forma que el terreno queda dividido en cuatro regiones distintas, que entonces estaban unidas mediante siete puentes llamados puente del herrero, Puente Conector, Puente Verde, Puente del Mercado, Puente de Madera, Puente Alto y Puente de la Miel. El problema se formul√≥ en el siglo XVIII y consist√≠a en encontrar un recorrido para cruzar a pie toda la ciudad pasando solo una vez por cada uno de los puentes y regresando al mismo punto de inicio.

El problema, formulado originalmente de manera informal, consist√≠a en responder a la siguiente pregunta: 

        "Dado el mapa de K√∂nigsberg, con el r√≠o Pregel dividiendo el plano en cuatro regiones distintas, que est√°n unidas a trav√©s de los siete puentes, ¬øes posible dar un paseo comenzando desde cualquiera de estas regiones, pasando por todos los puentes, recorriendo solo una vez cada uno, y regresando al mismo punto de partida? #### Solucion de Euler

Para dicha demostraci√≥n, Euler recurre a una abstracci√≥n del mapa y se enfoca exclusivamente en las regiones terrestres y las conexiones entre ellas. Cada puente qued√≥ representado mediante una l√≠nea que un√≠a a dos puntos, y cada uno de estos puntos representaba una regi√≥n diferente. As√≠, el problema se reduce a decidir si existe o no un camino que comience por uno de los puntos, recorra todas las l√≠neas una sola vez y regrese al mismo punto de partida. 



Euler determin√≥, en el contexto del problema, que los puntos intermedios de un recorrido posible necesariamente han de estar conectados a un n√∫mero par de l√≠neas. En efecto, si llegamos a un punto desde alguna l√≠nea, entonces el √∫nico modo de salir de ese punto es por una l√≠nea diferente. Esto significa que tanto el punto inicial como el final ser√≠an los √∫nicos que podr√≠an estar conectados con un n√∫mero impar de l√≠neas. Sin embargo, el requisito adicional del problema dice que el punto inicial debe ser igual al final, por lo que no podr√≠a existir ning√∫n punto conectado con un n√∫mero impar de l√≠neas.

En particular, como se ve en este diagrama, los cuatro puntos tienen un n√∫mero impar de l√≠neas (tres de ellos tienen tres l√≠neas y el restante tiene cinco). Por lo tanto, se concluye que es imposible definir un camino con las caracter√≠sticas buscadas que son los siete puentes de K√∂nigsberg.

# Algoritmo de Kamada Kuwai- M√©todo gr√°fico de fuerzas


Kamada y Kawai proponen en 1989 un m√©todo gr√°fico el cual tiene como finalidad realizar representaciones gr√°ficas de estructuras computacionales complejas entre las que podemos encontrar redes de Petri y diagramas de estado. Para realizar estas representaciones gr√°ficas, Kamada y Kawai proponen un algoritmo el cual est√° basado en el m√©todo de fuerzas.


Los algoritmos que se basan en el m√©todo de las fuerzas, llevan a cabo la gr√°fica de los par√°metros deseados usando la informaci√≥n contenida y despreciando la informaci√≥n de la gr√°fica inicial; Lo anterior es completamente opuesto al algoritmo
de Kruskal que tiene como objetivo buscar el sub grafo contenido dentro del grafo principal.


Para los algoritmos de fuerza al conjunto de nodos y v√©rtices que conforman la gr√°fica se le otorga un valor de energ√≠a, este valor es la energ√≠a acumulada del sistema. Los sistemas con un nivel de energ√≠a baja corresponden a nodos subyacentes que cuentan con distancias especificadas y que se encuentran cerca unos de otros, los nodos no subyacentes se encuentran a distancias muy grandes.


El algoritmo de Kamada- Kawai funciona tomando en cuenta las distancias m√°s cortas entre los v√©rtices. Este algoritmo simula fuerzas de resortes, tomando como base la ley de Hook, asignando un valor de fuerza dependiendo de la distancia te√≥rica y relativa entre los nodos que componen al sistema. Los pasos que utiliza el algoritmo de Kamada- Kawai son:

    1.- Se consideran la distancia Euclidiana entre los nodos
    2.- Se considera un sistema virtual en el que cada nodo se encuentra unido mediante un resorte.
    3.- Posteriormente se posicionan los nodos de manera que el estado total de la energ√≠a en el sistema de resortes sea m√≠nimo.

Por tanto, las fuerzas asociadas a los muelles son tanto atractivas como repulsivas para conseguir que la localizaci√≥n de los nodos respete las distancias te√≥ricas que existen en el grafo

 ‚Ä¢$ \ p_{i}= (x_{i} , y_{i} ) $ y $ \ p_{j}= (x_{j} , y_{j} ) $ son las posiciones de los nodos i y j en el grafico

 ‚Ä¢$ \ d(p_{i},p_{j})= |p_{i}- p_{j}| $ es la distancia Euclidea entre dichas posiciones

 ‚Ä¢$ \ l_{i,j} $ es la distancia te√≥rica en el grafo entre dichos nodos 

	
$ \mathbf{Objetivo:} $  Encontrar una distribuci√≥n tal que para cada par de nodos i y j, d(pi,pj) sea aproximadamente igual a li,j

 ‚Ä¢ Es decir, el sistema tiene una fuerza proporcional a d(pi,pj) - li,j con la siguiente funcion de energ√≠a global, que se debe minimizar:
 
 E= $ 	\sum_{i=1}^{n-1} \sum_{j=i+1}^{n} \frac{1}{2} k_{i,j} (|p_{i} - p_{j} | - l_{i,j})^{2} $ 
 
 
 ‚Ä¢ La energ√≠a potencial en el muelle que conecta i y j es: $\ ¬Ω ¬∑ k_{ij} ¬∑ (d(p_{i},p_{j}) -l_{i,j})^{2} $

$ \Delta_{m}= \sqrt{(\frac{\delta E}{\delta x_{m}})^{2}+(\frac{\delta E}{\delta y_{m}})^{2}} $
